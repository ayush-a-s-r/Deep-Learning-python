```python
from statsmodels.tsa import seasonal,arima_model
import statsmodels.api as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the dataset
data=pd.read_csv('/content/monthly-milk-production-pounds.csv')
```

```python
#data cleaning
data.dropna(axis=0,inplace=True)
```

```python
#column name change
data.columns = ['mon','mp']
```

```python
#set datetime as index
data['mon']=pd.to_datetime(data['mon'])
```

```python
data.set_index('mon',inplace=True)
```

```python
tm=data['mp']
```

```python
tm1=tm.copy()
```

```python
vd=seasonal.seasonal_decompose(tm1)
```

```python
vd.plot()
```

```python
num_sample_to_forecast=24
```

```python
for i in range(0,num_sample_to_forecast):
  #seasonal model creation for arima
  model1 = st.tsa.statespace.SARIMAX(tm,order=(0,1,1),seasonal_order=(1,1,1,12))
  res1=model1.fit() #training
  new_op=res1.forecast() #forecasting only one value
  print(new_op)
  tm=pd.concat([tm,new_op]) #train data concatenation

tm.plot()
tm1.plot()
plt.show()
```
